```python
import matplotlib.pyplot as plt

swap_ipc1  = [1.22, 1.37, 2.17, 1.80, 1.94, 1.91, 1.99, 1.89, 2.05]   
swap_ipc10 = [6.32, 7.91, 8.26, 8.20, 7.23, 8.12, 8.19, 8.25, 8.38]  
swap_ipc50 = [12.77,20.14,19.69,20.71,19.94,20.28,19.83,21.07,20.93]


base_ipc1  = 2.2  
base_ipc10 = 8.3   
base_ipc50 = 21  

ks_ipc1  = [1, 2, 5, 10, 20, 50, 100, 150, 200]
ks_ipc10 = [1, 2, 5, 10, 20, 50, 100, 150, 200]  
ks_ipc50 = [1, 2, 5, 10, 20, 50, 100, 150, 200]


rel1  = [a/base_ipc1  * 100 for a in swap_ipc1]
rel10 = [a/base_ipc10 * 100 for a in swap_ipc10]
rel50 = [a/base_ipc50 * 100 for a in swap_ipc50]
plt.figure(figsize=(6,4))
plt.plot(ks_ipc1,  rel1,  'o-', label='IPC=1')
plt.plot(ks_ipc10, rel10, 's--', label='IPC=10')
plt.plot(ks_ipc50, rel50, 'd-.', label='IPC=50')
plt.xscale('log', base=2)
plt.xlabel('Swap i-th label (log scale)')
plt.ylabel('Relative performance (%)')
plt.title('TinyImageNet: i-th Label Swap Test') 
plt.legend()
plt.grid(True, which='both', ls=':')
plt.tight_layout()
plt.show()
```
